#### To get location and cardinal direction of each clients

```python
import pandas as pd
from geopy.geocoders import Nominatim
import math

# Initialize geolocator
geolocator = Nominatim(user_agent="cluster_locator")

# Reference point (can be your central hub or (0,0))
ref_lat, ref_lon = 0, 0

# Function to get location name
def get_address(lat, lon):
    if pd.isnull(lat) or pd.isnull(lon):
        return None
    try:
        location = geolocator.reverse((lat, lon), exactly_one=True, timeout=10)
        return location.address if location else None
    except:
        return None

# Function to calculate bearing
def calculate_bearing(lat1, lon1, lat2, lon2):
    lat1, lon1, lat2, lon2 = map(math.radians, [lat1, lon1, lat2, lon2])
    d_lon = lon2 - lon1
    x = math.sin(d_lon) * math.cos(lat2)
    y = math.cos(lat1) * math.sin(lat2) - math.sin(lat1) * math.cos(lat2) * math.cos(d_lon)
    bearing = math.atan2(x, y)
    bearing = math.degrees(bearing)
    return (bearing + 360) % 360

# Map bearing to 16-point compass
def bearing_to_cardinal(bearing):
    directions = [
        "North", "North-NorthEast", "NorthEast", "East-NorthEast", "East", "East-SouthEast", "SouthEast", "South-SouthEast",
        "South", "South-SouthWest", "SouthWest", "West-SouthWest", "West", "West-NorthWest", "NorthWest", "North-NorthWest"
    ]
    idx = int((bearing + 11.25) % 360 / 22.5)
    return directions[idx]

# Function to get cardinal direction from lat/lon
def get_cardinal_direction(lat, lon):
    if pd.isnull(lat) or pd.isnull(lon):
        return None
    bearing = calculate_bearing(ref_lat, ref_lon, lat, lon)
    return bearing_to_cardinal(bearing)

# Apply both functions separately to centers DataFrame
centers['location_name'] = centers.apply(
    lambda row: get_address(row['center_lat'], row['center_lon']), axis=1
)

centers['clients_direction'] = centers.apply(
    lambda row: get_cardinal_direction(row['center_lat'], row['center_lon']), axis=1
)

# Merge back to main df
df = df.merge(
    centers[['region_id', 'location_name', 'clients_direction']],
    on='region_id',
    how='left'
)
```

```python
# fix multi colinearity
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant

#  Define your numeric features
features = ['loannumber', 'loanamount', 'totaldue', 'termdays', 'age']

#  Drop rows with NaNs
X_original = df[features].dropna()

#  Compute correlation matrix BEFORE VIF filtering
corr_before = X_original.corr().abs()

#  VIF filtering function
def calculate_vif(X):
    X_const = add_constant(X)
    vif_data = pd.DataFrame()
    vif_data["feature"] = X_const.columns
    vif_data["VIF"] = [variance_inflation_factor(X_const.values, i) for i in range(X_const.shape[1])]
    return vif_data[vif_data["feature"] != "const"]

#  Iteratively drop features with VIF > 10
X_filtered = X_original.copy()
vif_data = calculate_vif(X_filtered)

while vif_data["VIF"].max() > 10:
    drop_feature = vif_data.sort_values("VIF", ascending=False).iloc[0]["feature"]
    print(f"Dropping {drop_feature} (VIF={vif_data['VIF'].max():.2f})")
    X_filtered = X_filtered.drop(columns=[drop_feature])
    vif_data = calculate_vif(X_filtered)

#  Compute correlation matrix AFTER VIF filtering
corr_after = X_filtered.corr().abs()

#  Plot side-by-side heatmaps
plt.figure(figsize=(16, 6))

# Left: Before VIF filtering
plt.subplot(1, 2, 1)
sns.heatmap(corr_before, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix Before VIF Filtering")
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)

# Right: After VIF filtering
plt.subplot(1, 2, 2)
sns.heatmap(corr_after, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix After VIF Filtering")
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)

plt.tight_layout()
plt.show()
```

#### Correlation matrix for categorical features using Cramér’s V -- This matrix suggests that while some demographic features like education and employment status have moderate influence on banking behavior, most variables are weakly correlated, indicating they capture distinct aspects of client profiles.

```python
from scipy.stats import chi2_contingency

# Function to compute Cramér's V
def cramers_v(x, y):
    confusion_matrix = pd.crosstab(x, y)
    chi2 = chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    k = min(confusion_matrix.shape)  # min(num_rows, num_cols)
    return np.sqrt(chi2 / (n * (k - 1)))

# Compute correlation matrix for categorical columns
corr_matrix = pd.DataFrame(
    np.zeros((len(cat_cols), len(cat_cols))),
    index=cat_cols,
    columns=cat_cols
)

for col1 in cat_cols:
    for col2 in cat_cols:
        corr_matrix.loc[col1, col2] = cramers_v(df[col1], df[col2])

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Cramér's V Correlation Between Categorical Features")
plt.show()

corr_matrix
```
